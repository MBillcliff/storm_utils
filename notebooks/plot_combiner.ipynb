```python
figure_dir = paths['storm_utils_figures'] / 'data_plots'

img1 = mpimg.imread(figure_dir / 'quadrant_TP.png')
img2 = mpimg.imread(figure_dir / 'quadrant_FP.png')
img3 = mpimg.imread(figure_dir / 'quadrant_FN.png')
img4 = mpimg.imread(figure_dir / 'quadrant_TN.png')

fig, axs = plt.subplots(2, 2, figsize=(18, 9))
# Remove all subplot spacing
plt.subplots_adjust(left=0, right=1, top=1, bottom=0,
                    wspace=0.0, hspace=0.0)

ax1, ax2, ax3, ax4 = axs.flat

# Show each image with correct aspect ratio
for ax, img in zip([ax1, ax2, ax3, ax4], [img1, img2, img3, img4]):
    ax.imshow(img)
    ax.set_aspect('equal')   # preserve native aspect ratio
    ax.axis("off")

# Panel labels
labels = ["a)", "b)", "c)", "d)"]
for ax, lab in zip([ax1, ax2, ax3, ax4], labels):
    ax.text(-0.05, 1.02, lab,
            transform=ax.transAxes,
            fontsize=26, fontweight="bold", color="black",
            va="top", ha="left", fontname="Arial")

# plt.tight_layout()
plt.savefig(figure_dir / "quadrant.png",
            bbox_inches="tight", pad_inches=0)
plt.show()
```

```python
from matplotlib.lines import Line2D
from matplotlib.patches import Patch

figure_dir = paths['storm_utils_figures'] / 'data_plots'

img1 = mpimg.imread(figure_dir / 'quadrant_TP.png')
img2 = mpimg.imread(figure_dir / 'quadrant_FP.png')
img3 = mpimg.imread(figure_dir / 'quadrant_FN.png')
img4 = mpimg.imread(figure_dir / 'quadrant_TN.png')

# define custom legend items matching your plots
legend_elements = [
    Line2D([0], [0], color="blue", lw=2, label="v ensemble"),
    Line2D([0], [0], color="k", lw=2, label="OMNI"),
    Line2D([0], [0], color="#E66100", lw=2, label="Hp30"),
    Line2D([0], [0], color="#D41159", lw=2, linestyle="--", label="Storm threshold"),
    Line2D([0], [0], color="k", lw=2, linestyle="--", label="$T_0$"),
    Patch(facecolor="grey", alpha=0.4, label="Forecast window"),
    Line2D([0], [0], marker="o", color="royalblue", markersize=10, linestyle="None", 
           label=r"$\mathrm{Hp30}_{\mathrm{MAX}}$")
]


fig, axs = plt.subplots(2, 2, figsize=(14, 10))
plt.subplots_adjust(left=0, right=1, top=1, bottom=0,
                    wspace=0.0, hspace=0.0)

for ax, img in zip(axs.flat, [img1, img2, img3, img4]):
    ax.imshow(img)
    ax.axis("off")

# Panel labels
labels = ["TP)", "FP)", "FN)", "TN)"]
for ax, lab in zip(axs.flat, labels):
    ax.text(-0.07, 1.02, lab,
            transform=ax.transAxes,
            fontsize=24, fontweight="bold", color="black",
            va="top", ha="left", fontname="Arial")


# Add one global legend at bottom
fig.legend(handles=legend_elements, loc="lower center", ncol=4, fontsize=20, bbox_to_anchor=(0.5, -0.15))

plt.savefig(figure_dir / "quadrant.png", bbox_inches="tight")
plt.show()
```
