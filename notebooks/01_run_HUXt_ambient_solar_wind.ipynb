## Running HUXt

This will take about 4 hours with 2000 ensembles. 

```python
import storm_utils.huxt_utils as HU
from sunpy.coordinates import sun
import datetime

# HUXt setup
start_cr = int(sun.carrington_rotation_number(datetime.datetime(1995, 1, 1))) + 1
end_cr = int(sun.carrington_rotation_number(datetime.datetime(2026, 1, 1))) - 1
n_crs = end_cr - start_cr

Nens = 2000
seed = 42
huxt_run_id = 3
OVERWRITE = True

print(f"{'='*60}")
print(f"HUXt Ensemble Generation Configuration")
print(f"{'='*60}")
print(f"Carrington Rotations: {start_cr} to {end_cr} ({n_crs} total)")
print(f"Time span: ~{n_crs * 27.25:.0f} days (~{n_crs * 27.25 / 365:.1f} years)")
print(f"Ensemble members: {Nens}")
print(f"Random seed: {seed}")
print(f"Run ID: {huxt_run_id}")
print(f"Overwrite existing: {OVERWRITE}")
print(f"{'='*60}\n")

# Estimate time and storage
est_time_per_cr = 30  # seconds (adjust based on your system)
est_total_hours = (n_crs * est_time_per_cr) / 3600
est_storage_per_cr = 50  # MB (for 2000 ensembles)
est_total_gb = (n_crs * est_storage_per_cr) / 1024

print(f"Estimated runtime: {est_total_hours:.1f} hours")
print(f"Estimated storage: {est_total_gb:.1f} GB")

# Confirm before running
response = input("\nProceed with HUXt run? (yes/no): ")
if response.lower() != 'yes':
    print("Run cancelled")
else:
    HU.run_multiple_ambient_ensembles(
        start_cr, n_crs, Nens, huxt_run_id, 
        seed=seed, overwrite=OVERWRITE
    )
```
